### 2. Clustering k-Means et k-Medoids 

#### 2.1 Pour démarrer

```python
def k_means(data,f0,f1,k,plot=False):
    print("Appel KMeans pour une valeur fixee de k")
    tps1 = time.time()
    model = cluster.KMeans(n_clusters=k, init='k-means++')
    model.fit(data)
    tps2 = time.time()
    labels = model.labels_
    iteration = model.n_iter_
    
    print("Coefficient de silhouette pour k =",k,":",metrics.silhouette_score(data,labels))
    print("Coefficient de Davies-Bouldin pour k =",k,":",metrics.davies_bouldin_score(data,labels))
    if(plot):
        plt.scatter(f0, f1, c=labels, s=8)
        plt.title("Donnees apres clustering Kmeans")
        plt.show()
    
    print("nv clusters =",k,", nb iter =", iteration,", ...... runtime = ", round((tps2-tps1)*1000,2),"ms")

print("DATA : \n")
    
for i in range(2,6):
    if(i == 2):
        k_means(data, f0,f1,i,True)
    else:
        k_means(data, f0,f1,i)
```

#### 2.4 Méthode k-medoids

```python
def k_medoids(data,f0,f1, k, distance_function=euclidean_distances):
    tps1 = time.time()
    distmatrix = distance_function(data)
    fp = kmedoids.fasterpam(distmatrix, k)
    tps2 = time.time()
    iter_kmed = fp.n_iter
    labels_kmed = fp.labels
    print('Loss with FasterPAM:', fp.loss)
    
    print("Coefficient de silhouette pour k =",k,":",metrics.silhouette_score(data,labels_kmed))
    print("Coefficient de Davies-Bouldin pour k =",k,":",metrics.davies_bouldin_score(data,labels_kmed))
    #plt.scatter(f0,f1, c=labels_kmed, s=8)
    #plt.title("Données apres clustering KMedoids")
    #plt.show()
    print(" nb clusters =",k,", nb_iter =",iter_kmed,", ...... runtime =", round((tps2-tps1)*1000,2),"ms")

databrut = arff.loadarff(open(path+"xclara.arff", 'r'))
data = [[x[0],x[1]] for x in databrut[0]]

#Affichage 2D
f0 = [f[0] for f in data]
f1 = [f[1] for f in data]

for i in range(2,10):
    k_medoids(data, f0,f1,i)
```

- le bon nombre de clusters permettant d'obtenir un coefficient de silhouette le plus proche de 1 est k = 3 pour cette même valeur nous avons un coefficient de Davies-Bouldin le plus faible signifiant un clustering qui est le plus optimisé. 
- Le temps de calcul pour une telle execution est 100,12 ms
- Non, on ne connais pas le résultat attendu donc on ne peux pas comparer. 

```python
def evaluation(data,f0,f1,k):
    
    # méthode K-means 
    model = cluster.KMeans(n_clusters=k, init='k-means++')
    model.fit(data)
    labels = model.labels_
    
    # méthode K-medoids
    distmatrix = euclidean_distances(data)
    fp = kmedoids.fasterpam(distmatrix, k)
    labels_kmed = fp.labels
    
    print("Similarity score of the data : \n\t- Rand_Score :",metrics.rand_score(labels,labels_kmed),"\n\t- Mutual_Info_Score :",metrics.mutual_info_score(labels,labels_kmed))
    
```

```python
# Nous reprenons le premier dataset :

databrut = arff.loadarff(open(path+"xclara.arff", 'r'))
data = [[x[0],x[1]] for x in databrut[0]]

#Affichage 2D
f0 = [f[0] for f in data]
f1 = [f[1] for f in data]

evaluation(data,f0,f1,3)
```

### 3. Clustering agglomératif

#### 3.1 Pour démarrer

```python
def print_dendrogramme(data):
    # Donnees dans datanp
    print (" Dendrogramme ’ single ’ donnees initiales")
    linked_mat = shc.linkage(data,'single')
    plt.figure(figsize = (12,12))
    shc.dendrogram(linked_mat,
    orientation = 'top',
    distance_sort = 'descending',
    show_leaf_counts = False)
    plt.show()
    

print_dendrogramme(data)
```

- code permettant de déterminer un clustering hiérarchique en utilisant soit une limite sur le seuil de distance soit un nombre de clusters.

```python
def lim_cluster(data):
    # set distance_threshold ( 0 ensures we compute the full tree )
    tps1 = time.time()
    model = cluster.AgglomerativeClustering(distance_threshold = 7 , linkage = 'single' , n_clusters = None)
    model = model.fit(data)
    tps2 = time.time()
    labels = model.labels_
    k = model.n_clusters_
    leaves = model.n_leaves_
    # Affichage clustering
    plt.scatter(f0,f1,c = labels,s = 8 )
    plt.title(" Resultat du clustering ")
    plt.show()
    print(" nb clusters = ",k," , nb feuilles = ",leaves, " runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms " )

    #set the number of clusters 
    k = 4
    tps1 = time.time()
    model = cluster.AgglomerativeClustering(linkage = 'single' , n_clusters = k)
    model = model.fit(data)
    tps2 = time.time()
    
    labels = model.labels_
    k = model.n_clusters_
    leaves = model.n_leaves_
    # Affichage clustering
    plt.scatter(f0,f1,c = labels,s = 8 )
    plt.title(" Resultat du clustering ")
    plt.show()
    print(" nb clusters = ",k," , nb feuilles = ",leaves, " runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms " )
    
    
lim_cluster(data)
```

```python
def set_distance(data,f0,f1, distance, plot=False):
    # set distance_threshold ( 0 ensures we compute the full tree )
    tps1 = time.time()
    model = cluster.AgglomerativeClustering(distance_threshold = distance , linkage = 'single' , n_clusters = None)
    model = model.fit(data)
    tps2 = time.time()
    labels = model.labels_
    k = model.n_clusters_
    leaves = model.n_leaves_
    
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
    print("Coefficient de silhouette pour distance =",distance,":",metrics.silhouette_score(data,labels))
    print("Coefficient de Davies-Bouldin pour distance =",distance,":",metrics.davies_bouldin_score(data,labels))
    print(" nb clusters = ",k," , nb feuilles = ",leaves, " runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms " , "for a distance of :", distance)
    print("//////////////////////////////////////////////////////////////")
for i in range (1,10,1):
    if(i == 1): set_distance(data3,f2,f3,i/1,True)
    else : set_distance(data3,f2,f3,i/1)
        


```

```python
# Pour le second dataset nous avons :
for i in range (1,10,1):
    if(i == 1): set_distance(data4,f4,f5,i/10,True)
    else : set_distance(data4,f4,f5,i/10)
```

- Pour les distances actuelles nous allons voir les differentes manières de combiner des clusters.

```python
def set_linkage(data,f0,f1, distance, link, plot=False):
    # set distance_threshold ( 0 ensures we compute the full tree )
    tps1 = time.time()
    model = cluster.AgglomerativeClustering(distance_threshold = distance , linkage = link , n_clusters = None)
    model = model.fit(data)
    tps2 = time.time()
    labels = model.labels_
    k = model.n_clusters_
    leaves = model.n_leaves_
    
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
    print("Pour le linkage :", link)
    print("Silhouette :",metrics.silhouette_score(data,labels))
    print("Davies-Bouldin :",metrics.davies_bouldin_score(data,labels))
    print("nb clusters = ",k," , nb feuilles = ",leaves, " runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms ")
    print("//////////////////////////////////////////////////////////////")
```

```python
#Pour le premier dataset que nous avons choisi avec la valeur de distance 5.0 :

set_linkage(data3,f2,f3,5.0,'single',True)
set_linkage(data3,f2,f3,5.0,'average',True)
set_linkage(data3,f2,f3,5.0,'complete',True)
set_linkage(data3,f2,f3,5.0,'ward',True)
```

- Le meilleur linkage pour le premier dataset reste le single dans ce cas.

```python
#Pour le second dataset que nous avons choisi avec la valeur de distance 0.6 :

set_linkage(data4,f4,f5,0.6,'single',True)
set_linkage(data4,f4,f5,0.6,'average',True)
set_linkage(data4,f4,f5,0.6,'complete',True)
set_linkage(data4,f4,f5,0.6,'ward',True)
```

- Dans ce cas la, le linkage single reste toujours celui qui apporte des valeurs les plus bonnes.

- Les différentes valeurs de temps d'execution sont pour le premier dataset :

```python
# Changement de la valeur du nombre de clusters. 
def set_nb_clusters(data,f0,f1, nb_clusters, plot=False):
    #set the number of clusters 
    k = nb_clusters
    tps1 = time.time()
    model = cluster.AgglomerativeClustering(linkage = 'single' , n_clusters = k)
    model = model.fit(data)
    tps2 = time.time()
    
    labels = model.labels_
    k = model.n_clusters_
    leaves = model.n_leaves_
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
        
    print("Nb clusters = ",k," , nb feuilles = ",leaves, " runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms " )
    print("Silhouette :",metrics.silhouette_score(data,labels))
    print("Davies-Bouldin :",metrics.davies_bouldin_score(data,labels))
    print("//////////////////////////////////////////////////////////////")
    
```

### 4. Clustering DBSCAN et HDBSCAN 

#### 4.1 Intérêts de la méthode DBSCAN

- Pour cette partie nous allons utiliser pour dataset le dataset 3, ainsi que le dataset numéro 4 qui sont des clusters assez isolés donc facilement traitables 

```python
def DBSCAN(data,f0,f1,plot = False):  
    tps1 = time.time()
    model = cluster.DBSCAN(eps=1,min_samples= 2)
    model = model.fit(data)
    tps2 = time.time()
    
    
    labels = model.labels_
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
    
    print("runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms " )
    print("Silhouette :",metrics.silhouette_score(data,labels))
    print("Davies-Bouldin :",metrics.davies_bouldin_score(data,labels))
    print("//////////////////////////////////////////////////////////////")
```

```python
DBSCAN(data3,f2,f3,True)
DBSCAN(data4,f4,f5,True)
```

- Ici pour DBSCAN on initialise des valeurs qui sont aléatoires. Nous allos tout d'abord commencer par faire varier epsilon et ensuite nous vrrons les modifications 

```python
def DBSCAN_epsilon(data,f0,f1,epsilon,plot=False):
    tps1 = time.time()
    model = cluster.DBSCAN(eps=epsilon,min_samples= 2)
    model = model.fit(data)
    tps2 = time.time()
    
    
    labels = model.labels_
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
    
    print("runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms pour un epsilon égal à ", epsilon )
    print("Silhouette :",metrics.silhouette_score(data,labels))
    print("Davies-Bouldin :",metrics.davies_bouldin_score(data,labels))
    print("//////////////////////////////////////////////////////////////")
```

```python
for i in range(1,10,1):
    DBSCAN_epsilon(data3,f2,f3,i,True)
```

- le resultat permettant d'avoir des valeurs qui semblent correctes pour les coefficient à une valeur de epsilon de 5. Nous allons maintenant essayer de voir si nous pouvons affiner ce résultat en faisant varier min_samples. 

```python
def DBSCAN_samples(data,f0,f1,samples,epsilon=5,plot=False):
    tps1 = time.time()
    model = cluster.DBSCAN(eps=epsilon,min_samples= samples)
    model = model.fit(data)
    tps2 = time.time()
    
    
    labels = model.labels_
    # Affichage clustering
    if(plot):
        plt.scatter(f0,f1,c = labels,s = 8 )
        plt.title(" Resultat du clustering ")
        plt.show()
    
    print("runtime = " ,round((tps2 - tps1) * 1000 , 2 ) ," ms pour une valeur de min_samples égale à ", samples )
    print("Silhouette :",metrics.silhouette_score(data,labels))
    print("Davies-Bouldin :",metrics.davies_bouldin_score(data,labels))
    print("//////////////////////////////////////////////////////////////")
```

```python
for i in range(1,20):
    DBSCAN_samples(data3,f2,f3,i,5,True)
```

- Dans notre cas, la valeur de min_sample qui fonctionne le mieux est 1.


- Nous allons suivre lemême processus pour le deuxième dataset que nous avons décidé de traiter

```python
for i in range(1,20,1):
    DBSCAN_epsilon(data4,f4,f5,i/10,True)
```

- La meilleure valeur de epsilon que l'on obtient dans ce cas est 0.6 maintenant intéressons nous a min_samples

```python
for i in range(1,20):
    DBSCAN_samples(data4,f4,f5,i,0.6,True)
```

- les valeurs pour les coefficient semblent diminuer lorsque l'on augmente la valeur de min_sample donc en s'en tenant à 1 cela fonctionne.
